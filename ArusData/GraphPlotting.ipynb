```python
import pandas as pd
import base64
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/Lo/ArusData/ProgramOverallData.csv")
```

```python
#print(df['T22-Program ini sesuai dengan keperluan perkembangan profesional saya. '].info())
mean_x = []
Program = []
for i in df['WHat'].unique():
        #print(x)
        condition = df['WHat'] == i
        Program.append(i)
        mean_x.append(df[condition]['T15-Saya dapat memahami dan mengikuti sesi ini. '].mean())
Frame = pd.DataFrame(list(zip(Program,mean_x)),columns =['Program', 'Mean'])
#display(Frame)

Frame.plot(kind = "bar",x='Program' , y="Mean")
plt.xticks(rotation=90)
plt.show()
#print(max(mean_x))
print(max(Frame['Mean']))
Frame.max()
```
